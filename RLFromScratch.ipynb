```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Chargement des données
df = pd.read_csv('social_media_productivity_6000.csv').dropna()

# 2. Définition des variables (Vecteurs)
x = df['study_hours'].values # le variable qui presente les heurs de study
y = df['productivity_score'].values # le variable qui presente le score de prodectivite

# 3.calcule le centre de gravité des données signifie que la droite de régression passe obligatoirement par le point ( x_bar et y_bar) ( c'est la moyenne)
x_mean = np.mean(x)
y_mean = np.mean(y)

# 4. CALCUL DE LA PENTE (a)
# Numérateur : Somme de (x - x_mean) * (y - y_mean) (la covariance => Si x augmenter quand y augmenté donc le résultat sera positif.)
numerator = np.sum((x - x_mean) * (y - y_mean))
# Dénominateur : Somme de (x - x_mean)^2 ( on calcule la variance ) 
denominator = np.sum((x - x_mean)**2)

a = numerator / denominator # L'impact precis d'une variable sur l'autre.

# 5. CALCUL DE L'INTERCEPTION (b)
b = y_mean - (a * x_mean)

print(f"Équation calculée : y = {a:.4f}x + {b:.4f}")

# 6. CALCUL DES PRÉDICTIONS
# On applique l'équation y = ax + b à chaque valeur de x
y_pred = a * x + b

# 7. VISUALISATION
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.3, label='Données réelles') # le nuage de points
plt.plot(x, y_pred, color='red', linewidth=2, label=f'Droite: y={a:.2f}x + {b:.2f}') # la droite de régression
plt.xlabel('Heures d\'étude') # l'axe d'heures d'etude
plt.ylabel('Productivité') # l'axe de productivite
plt.legend()
plt.title('Régression Linéaire : Calculs Manuels')
plt.show()
```
