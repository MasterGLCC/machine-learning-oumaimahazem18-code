```python
import pandas as pd # Pandas est une bibliothèque logicielle pour tranformer les fichiers bruts en une structure DataFrame
import numpy as np # numpy est une bibliothèque standard pour les calculs mathématiques et scientifiques en Python
import matplotlib.pyplot as plt # c'est une bibliothèque de traçage qui permet de générer des graphiques de qualité professionnelle
from sklearn.linear_model import LinearRegression # bibliothèque de référence en Python qui ontient des centaines d'algorithmes prêts à l'emploi pour la regression
from sklearn.metrics import mean_squared_error, r2_score # est un outile de mesure les erreurs utilise pour teste la performance de notre modele

# Chargement des données
df = pd.read_csv('social_media_productivity_6000.csv')# cette ligne permet de lire DataSet

# Nettoyage rapide (suppression des lignes avec des valeurs manquantes)
df = df.dropna()
```

```python
# Variables : X (indépendante), y (dépendante)
X_simple = df[['study_hours']]
y = df['productivity_score']

# Création et entraînement du modèle
model_simple = LinearRegression() # Le modèle ne connaît pas encore les données comme une boîte à outils mathématiques vide
model_simple.fit(X_simple, y) # entrainement du modele

# Visualisation
plt.scatter(X_simple, y, color='blue', alpha=0.3, s=10) #dessiner le nuage de points initial,alpha=0.3(les points sont semi-transparents),s=10(la taille de chaque points
plt.plot(X_simple, model_simple.predict(X_simple), color='red', linewidth=2) # tracer la ligne de régression
plt.xlabel('Heures d\'étude') # donner un nom à l'axe horizontal
plt.ylabel('Score de productivité') # nommer l'axe vertical
plt.title('Régression Linéaire Simple') # le titre du graphique
plt.show() #afficher le graphique

print(f"R² Score: {model_simple.score(X_simple, y):.4f}") # utilisée pour calculer et afficher la précision de notre modèle (si le droite de régression est une bonne approximation de la réalité ou non)
```
